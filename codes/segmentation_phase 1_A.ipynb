```python
import re
from pathlib import Path
import pandas as pd

data_folder = Path("C:/Users/Jing Li/Desktop/cv2db/project/Inputs")
data_to_open = data_folder / "Resume_A.txt"
```

```python
with open(data_to_open) as f: 
	data = f.read()
    
# Seperate each person
pattern = r"Qualifications and Experience"
temp = list(filter(None, re.split(pattern, data)))
del temp[0]

# seperate QE

pattern_QE = r"\n\d+\. "
# QE = [re.split(pattern_QE, i) for i in temp]
QEs = [re.split(pattern_QE, i)[0] for i in temp] # QEs in the form of a list
data = [re.split(pattern_QE, i)[1] for i in temp] # information without QE, a list
```

```python
#Checking
# print(len(QE), QE)
# print(len(data), data[0])
```

```python
# Split: name and title

name_title = [i.split("\n", 1)[0] for i in data]
ppls_noname = [i.split("\n", 1)[1] for i in data]

name_title = [i.replace(" Summary", "") for i in name_title]

pattern_name = r"\s[–-]\s"
name = [re.split(pattern_name, i)[0] for i in name_title]
title = [re.split(pattern_name, i)[1] for i in name_title]

# remove degree from name
name = [re.split(",", i)[0] for i in name]
```

```python
# Split: career summary, education

smmry = [re.split(r"\nEducation\n|Education, Training", ppl)[0] for ppl in ppls_noname]
ppls_nosmy = [re.split(r"\nEducation\n|Education, Training", ppl)[1] for ppl in ppls_noname]

edu_pattern = r"(Selected Project|Relevant Project|Work|Professional)\sExperience"

education = [re.split(edu_pattern, ppl, maxsplit = 1)[0] for ppl in ppls_nosmy]
ppls_noedu = [re.split(edu_pattern, ppl, maxsplit = 1)[2] for ppl in ppls_nosmy]
```

```python
# Split: Work Experience
# 16 people has additional materials after work experience

pattern_work = r"\n(Selected |Recent )?Publications.*|(\nSuccessful Grant |\sSelected )?Awards.*|.+ Presentations\n|\nConferences"

# Recent Publications and Presentations
# Publications
# Selected Publications and Technical Reports
# Selected Publications
# Awards and Professional Achievements|and then: Selected Publications
# Recent Technical Reports, and Presentations
# Publications, Technical Reports, and Presentations
# Selected Publications and Presentations
# Publications and Awards
# Publications and Presentations
# Awards
# Awards and Honors
# Successful Grant Awards
# Conferences
# Poster Presentations

ppls = [re.split(pattern_work, i, maxsplit = 1) for i in ppls_noedu]
print(ppls[1][1])
print([len(k) for k in ppls])
# for j in ppls:
#     print(len(j))
wrk = [i[0] for i in ppls]
ppls_others = [i[-1] if len(i) != 1 else None for i in ppls]
```
